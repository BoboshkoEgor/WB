```python
df_2 = pd.read_csv("Z:/WB/3_py/tarantino.csv")
df_2
```

```python
word_df=df_2[['movie', 'word', 'minutes_in']].loc[df_2['type']=='word']
rip_df=df_2[['movie', 'minutes_in']].loc[df_2['type']=='death']
len(word_df)
```

## Число проклятий и летальных исходов, отношение по ним в фильмах

```python
word_ac=pd.pivot_table(word_df, values = ['word'],
                      index=['movie'], aggfunc={'word': "count"})
word_ac['out_of_all']=round(word_ac['word']*100/len(word_df),2)
word_ac
```

```python
rip_ac=pd.pivot_table(rip_df, values = ['movie'],
                      index=['movie'], aggfunc={'movie': 'count'})
rip_ac['out_of_all']=round(rip_ac['movie']*100/len(rip_df),2)
rip_ac
```

## Сколько нужно ждать следующего ругательства при просмотре фильма

```python
word_df['wait_till_curse'] = 0.
word_df['is_new_movie']=(word_df['movie']!=word_df['movie'].shift(1))
```

```python
word_df.loc[word_df['is_new_movie']==True, 'wait_till_curse'] = word_df['minutes_in']
word_df.loc[word_df['is_new_movie']==False, 'wait_till_curse'] = word_df['minutes_in']-word_df['minutes_in'].shift(1)
word_df[['movie', 'word', 'wait_till_curse']]
```
